- Write a program to accept the price of a bike and display the road tax to be paid according to the following criteria :    

Price | Tax | 
------|-----|
> 100.000 | 15 % |
> 50.000 and <= 100.000 | 10 % 
<= 50000 | 5 %

- Given the array of numbers, write a program to move all zeros to the end keeping the relative order of the non-zero elements.

- Given two sentences, return an array that has the words that appear in one sentence and not the other and an array with the words in common.

```python
def bike_price():

    while True:
        try:
            price = int(input("Price of the bike: "))
            if price > 10000:
                price_tax = price * 1.15
                print(f"You will pay {price_tax}€ (tax included)")
            elif price > 50000 and price <= 10000:
                price_tax = price * 1.1
                print(f"You will pay {price_tax}€ (tax included)")
            if price <= 50000:
                price_tax = price * 1.05
                print(f"You will pay {price_tax}€ (tax included)")

        except ValueError:
            print("Invalid input. Please try again.")
            continue

        choice = input("Do you want to check another price (y/n): ")
        while choice.lower() not in ["y","n"]:
            choice = input("Please choose yes (y) or no (n): ")
        if choice.lower() == "n":
            print("Okay. Have a great day!")
            break

bike_price()
```

```python
phrase1 = "Hello, how are you? Good."
phrase2 = "Hi, how are you? Bad."

x = phrase1.split()
y = phrase2.split()

x = set(x)
y = set(y)

diff_x, diff_y = x-y, y-x
print(f"Words that differ: {diff_x, diff_y}")

common_words= x.intersection(y)
print(f"Common words: {common_words}")
```
